```python
def correlation_heatmap(df):
    _ , ax = plt.subplots(figsize =(14, 12))
    colormap = sns.diverging_palette(220, 10, as_cmap = True)
    
    _ = sns.heatmap(
        df.corr(), 
        cmap = colormap,
        square=True, 
        cbar_kws={'shrink':.9 }, 
        ax=ax,
        annot=True, 
        linewidths=0.1,vmax=1.0, linecolor='white',
        annot_kws={'fontsize':12 }
    )
    
    plt.title('Pearson Correlation of Features', y=1.05, size=15)

df_train['v1'] = df_train['v1'].map( {'a': 0, 'b': 1} ).astype(int)
df_test['v1'] = df_test['v1'].map( {'a': 0, 'b': 1} ).astype(int)
df_train['v4'] = df_train['v4'].map( {'u': 0, 'y': 1, 'l': 2} ).astype(int)
df_test['v4'] = df_test['v4'].map( {'u': 0, 'y': 1, 'l': 2} ).astype(int)
df_train['v6'] = df_train['v6'].map( {'c': 0, 'q': 1, 'W': 2, 'cc': 3, 'x': 4, 'aa': 5, 'i': 6, 'm': 7, 'k': 8, 'e': 9, 'ff': 10, 'd': 11, 'j': 12, 'r': 13} ).astype(int)
df_test['v6'] = df_test['v6'].map( {'c': 0, 'q': 1, 'W': 2, 'cc': 3, 'x': 4, 'aa': 5, 'i': 6, 'm': 7, 'k': 8, 'e': 9, 'ff': 10, 'd': 11, 'j': 12, 'r': 13} ).astype(int)
df_train['v7'] = df_train['v7'].map( {'v': 0, 'h': 1, 'bb': 2, 'ff': 3, 'z': 4, 'j': 5, 'n': 6, 'dd': 7, 'o': 8} ).astype(int)
df_test['v7'] = df_test['v7'].map( {'v': 0, 'h': 1, 'bb': 2, 'ff': 3, 'z': 4, 'j': 5, 'n': 6, 'dd': 7, 'o': 8} ).astype(int)
df_train['v9'] = df_train['v9'].map( {'f': 0, 't': 1} ).astype(int)
df_test['v9'] = df_test['v9'].map( {'f': 0, 't': 1} ).astype(int)
df_train['v10'] = df_train['v10'].map( {'f': 0, 't': 1} ).astype(int)
df_test['v10'] = df_test['v10'].map( {'f': 0, 't': 1} ).astype(int)
df_train['v12'] = df_train['v12'].map( {'f': 0, 't': 1} ).astype(int)
df_test['v12'] = df_test['v12'].map( {'f': 0, 't': 1} ).astype(int)

correlation_heatmap(df_train)
correlation_heatmap(df_test)
```

```python
df_train['isV9'] = df_train['v9'] == df_train['classLabel']
df_test['isV9'] = df_test['v9'] == df_test['classLabel']
```

```python
print(df_train['isV9'].mean())
print(df_test['isV9'].mean())
```

```python
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier
```

```python
del df_train['isV9']
del df_test['isV9']
y_train = df_train['classLabel']
del df_train['classLabel']
y_test = df_test['classLabel']
del df_test['classLabel']
```

```python
model = XGBClassifier()
model.fit(df_train, y_train)
```

```python
y_pred = model.predict(df_test)
predictions = [round(value) for value in y_pred]
# evaluate predictions
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
